### Task 1.1 - Data Addition

Since the age are hard to analyze, we will create a new column of age groups, that eqaully divided the age column into 4 age groups.

```python
# Define bins and labels for the age groups
age_bins = [16, 27, 38, 49, 60]  # 4 groups with 5 bins
age_labels = ['16 to 26', '27 to 37', '38 to 48', '49 to 59']

# Add a new column for age groups
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Display the updated dataset
df.head()
```

```python
# check fro missing values again

df.isnull().sum()
```

Now, all columns have no missing values, including the age group new column.

### Question 3 - How do user behavior is influenced by gender and age group?

### Question 6 - How do screen on time, battery drain, and data usage correlate?

We will plot a scatterplot on screen time and battery drain and a scatterplot on screen time and data usage, and a scatter plot on battery drain and data usage.

We will use subplots in vertical alignment.

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 18))

# Plot 1: Screen On Time vs Battery Drain
sns.scatterplot(x='Screen On Time (hours/day)', y='Battery Drain (mAh/day)', data=df, ax=axes[0])
axes[0].set_title('Screen On Time vs Battery Drain')

# Plot 2: Screen On Time vs Data Usage
sns.scatterplot(x='Screen On Time (hours/day)', y='Data Usage (MB/day)', data=df, ax=axes[1])
axes[1].set_title('Screen On Time vs Data Usage')

# Plot 3: Battery Drain vs Data Usage
sns.scatterplot(x='Battery Drain (mAh/day)', y='Data Usage (MB/day)', data=df, ax=axes[2])
axes[2].set_title('Battery Drain vs Data Usage')
```

The battery drain, screen on time, and data usage are all positively and closely correlated.
